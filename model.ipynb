```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from itertools import combinations
```

```python
df_ml = pd.read_csv("df_ml.csv").reset_index()
```

```python
df_ml.sample(10)
```

```python
df_metrics = pd.read_csv("top_metric_groups.csv")
```

```python
models = {}
metrics = df_metrics['combo'][0].replace("(","").replace(")","").replace("'","").split(", ")

      # loop through each mectric combination
year_results = {}
    

    
df_rfr = df_ml[ metrics + ['wRC+', 'year']].dropna()
    

x_train = df_rfr[metrics][df_rfr['year'] < 2023]        # split training data
y_train = df_rfr['wRC+'][df_rfr['year'] < 2023]



rfr = RandomForestRegressor(n_estimators=50, max_depth=None, max_features=0.2, random_state=42)
rfr_model = rfr.fit(x_train, y_train)

    
```
